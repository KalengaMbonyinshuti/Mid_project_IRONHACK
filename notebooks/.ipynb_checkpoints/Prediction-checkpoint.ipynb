```python
data=pd.read_csv('../data/CLEAN/final_data.csv')
data
```

```python
mean_value_Q2=data['Q2 Balance'].mean()
data['Q2 Balance'].fillna(value=mean_value_Q2, inplace=True)

mean_value_Q3=data['Q3 Balance'].mean()
data['Q3 Balance'].fillna(value=mean_value_Q3, inplace=True)

mean_value_average=data['Average Balance'].mean()
data['Average Balance'].fillna(value=mean_value_average, inplace=True)
```

```python
numerical_columns=data.select_dtypes(exclude='object')
categorical_columns=data.select_dtypes(include='object')
```

```python
cat_encoded=pd.get_dummies(data=categorical_columns, drop_first=True)
cat_encoded
```

```python
data_final=cat_encoded.join(numerical_columns)
data_final.drop('Customer Number', inplace=True, axis=1)
```

```python
X=data_final.drop('Offer Accepted_Yes', axis=1)
y=data_final['Offer Accepted_Yes']
```

```python
# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=config['train']['test_size'],random_state=config['train']['random_state'])
```

```python
# Place all the columns in the same scale [0,1].
# Add the import of the MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

# Save this scaled into scalers/ folder-. Check out the last Firday slides. "Pickle"

X_train_scaled_np = scaler.transform(X_train)
X_test_scaled_np  = scaler.transform(X_test)

X_train_scaled_df = pd.DataFrame(X_train_scaled_np, columns=X_train.columns, index=X_train.index)
X_test_scaled_df  = pd.DataFrame(X_test_scaled_np,  columns=X_test.columns,  index=X_test.index)
```

```python
# Create an instance of the LogisticRegression class
model = LogisticRegression()

# Train the model on the input data and target variables
model.fit(X_train_scaled_df, y_train)

# Save this model into models/ folder. Check out the last Firday slides. "Pickle"

# Train prediction
Train_prediction=model.predict(X_train_scaled_df)
Train_prediction

# Test Prediction
Test_prediction=model.predict(X_test_scaled_df)
Test_prediction
```

```python
# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, Test_prediction)
print("Accuracy: ", round(accuracy,2))
```

```python
score = model.score(X_test_scaled_df, y_test)
print("Accuracy:", round(score,2))
```

```python
# Evaluate the accuracy of the model
accuracy = accuracy_score(y_train, Train_prediction)
print("Accuracy: ",  round(accuracy,2))
```

```python
kappa = cohen_kappa_score(y_test, Test_prediction)
kappa
```

# Kappa regularization

```python
# Split the data into features (X) and target (y)
X1=data_final.drop('Offer Accepted_Yes', axis=1)
y1=data_final['Offer Accepted_Yes']

# Split the data into training and testing sets
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=config['train']['test_size'],random_state=config['train']['random_state'])


# Place all the columns in the same scale [0,1].
# Add the import of the MinMaxScaler
scaler1 = MinMaxScaler()
scaler1.fit(X1_train)

# Save this scaled into scalers/ folder-. Check out the last Firday slides. "Pickle"

X_train_scaled_np_1 = scaler1.transform(X1_train)
X_test_scaled_np_1  = scaler1.transform(X1_test)

X_train_scaled_df = pd.DataFrame(X_train_scaled_np_1, columns=X1_train.columns, index=X_train.index)
X_test_scaled_df  = pd.DataFrame(X_test_scaled_np_1,  columns=X1_test.columns,  index=X_test.index)

# Perform undersampling on the training data
rus1 = RandomUnderSampler()
X1_resampled, y1_resampled = rus1.fit_resample(X1_train, y1_train)
```

```python
# Train the model on the resampled data
classifier1 = LogisticRegression(solver='saga', max_iter=350)
classifier1.fit(X1_resampled, y1_resampled)

# Make predictions on the test data
y_pred1 = classifier1.predict(X1_test)

# Evaluate the model using accuracy, precision, recall, F1 score, and kappa score
acc = accuracy_score(y1_test, y_pred1)
prec = precision_score(y1_test, y_pred1)
rec = recall_score(y_test, y_pred1)
f1 = f1_score(y1_test, y_pred1)
kappa = cohen_kappa_score(y1_test, y_pred1)
```

```python
print("Accuracy:", round(acc, 2))
print("Precision:", round(prec, 2))
print("Recall:", round(rec, 2))
print("F1 Score:", round(f1, 2))
print("Kappa Score:", round(kappa, 2))
```
