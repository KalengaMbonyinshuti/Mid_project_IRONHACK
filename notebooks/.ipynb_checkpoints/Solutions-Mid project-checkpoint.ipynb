What is the average balance of the customers grouped by Income Level? The returned result should have only two columns, income level and Average balance of the customers. Use an alias to change the name of the second column.

```python
# Group the data by income level and calculate the average balance for each group
grouped_Income_Level = data.groupby("Income Level")["Average Balance"].mean().reset_index()

# Rename the second column to Average_Balance
grouped_Income_Level = grouped_Income_Level.rename(columns={"Average Balance": "Average_Balance of the customers"})
grouped_Income_Level
```

With the above table, Customer's average balance is directly related to the customer's income level which simply means that the more increases the more average balance increases.

What is the average balance of the customers grouped by number_of_bank_accounts_open? The returned result should have only two columns, number_of_bank_accounts_open and Average balance of the customers. Use an alias to change the name of the second column.

```python
# Group the data by income level and calculate the average balance for each group
grouped_bank_accounts = data.groupby("# Bank Accounts Open")["Average Balance"].mean().reset_index()

# Rename the second column to Average_Balance
grouped_bank_accounts = grouped_bank_accounts.rename(columns={"Average Balance": "Average balance of the customers"})
grouped_bank_accounts
```

By analysis data,we can conclude that there is no direct relationship between number of accounts held with average balance as it can be shown in the table above since increase in Bank Accounts Open does not always reflect the increase in average balance.

9. Since this is something that the senior management is regularly interested in, create a view called `Customers__Balance_View1` of the same query.

```python
def Customers__Balance_View1(data):
    sorted_step_1=data[((data['Credit Rating']=='High')|(data['Credit Rating']=='Medium'))&(data['# Credit Cards Held']<=2)
    &(data['Household Size']>=3)&(data['Own Your Home']=="Yes")]
        
    customer_accepted_off=sorted_step_1[sorted_step_1['Offer Accepted']=='Yes']
    
    customer_accepted_off[customer_accepted_off["Average Balance"] < customer_accepted_off["Average Balance"].mean()]
    
```
